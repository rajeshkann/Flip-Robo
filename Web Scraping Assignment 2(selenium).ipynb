Q2: Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.

```python
naukri=webdriver.Chrome("chromedriver.exe")
naukri.get("https://www.naukri.com/")
```

```python
select1=naukri.find_element_by_id("qsb-keyword-sugg")
select1.clear()
select1.send_keys("Data Scientist")
time.sleep(4)
select2=naukri.find_element_by_id("qsb-location-sugg")
select2.clear()
select2.send_keys("Banglore")
time.sleep(4)
select3=naukri.find_element_by_class_name("btn")
select3.click()
```

```python
name=naukri.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
link=[]
for i in name:
    link.append(i.get_attribute("href"))
```

```python
job_description=[]
company=[]
job=[]
location=[]
for i in link[:10]:
    naukri.get(i)
    time.sleep(3)
    try:
        name=naukri.find_element_by_xpath("//h1[@class='jd-header-title']")
        job.append(name.text)
    except:
        name=naukri.find_element_by_xpath("//h1[@class='av-special-heading-tag ']")
        job.append(name.text)
    try:
        loc=naukri.find_element_by_xpath("//span[@class='location ']")
        location.append(loc.text)
    except:
        loc=naukri.find_element_by_xpath("//span[@class='slide-meta-loc pull-left']")
        location.append(loc.text)
    try:
        com=naukri.find_element_by_xpath("//a[@class='pad-rt-8']")
        company.append(com.text)
    except:
        com=naukri.find_element_by_xpath("//div[@class='f14 lh18 alignJ']")
        company.append(com.text)
    try:
        des=naukri.find_element_by_xpath("//section[@class='job-desc']")
        job_description.append(des.text)
    except NoSuchElementException:
        des=naukri.find_element_by_xpath("//section[@class='getJobDescription JD av_textblock_section']")
        job_description.append(des.text)
job_description
```

```python
jobs=pd.DataFrame({
    "Job":job,
    "company":company,
    "location":location,
    "job description":job_description
})
jobs
```

Q5: Write a python program to scrape the salary data for Data Scientist designation in Noida location.

```python
glass=webdriver.Chrome("chromedriver.exe")
glass.get("https://www.glassdoor.co.in/Salaries/index.htm")
```

```python
select1=glass.find_element_by_id("KeywordSearch")
select1.clear()
select1.send_keys("Data Scientist")
time.sleep(4)
select2=glass.find_element_by_id("LocationSearch")
select2.clear()
select2.send_keys("Noida")
time.sleep(4)
select3=glass.find_element_by_id("HeroSearchButton")
select3.click()
```

```python
company=[]
average=[]
salary=[]
mini=[]
maxi=[]
com=glass.find_elements_by_xpath("//p[@class='m-0 ']")
for i in com:
    company.append(i.text)
avg=glass.find_elements_by_xpath("//div[@class='col-2 d-none d-md-flex flex-row justify-content-end']")
for i in avg:
    average.append(i.text.replace("\n",""))
sal=glass.find_elements_by_xpath("//p[@class='css-1uyte9r css-1kuy7z7 m-0 ']")
for i in sal:
    salary.append(i.text)
minie=glass.find_elements_by_xpath("//div[@class='common__RangeBarStyle__values d-flex justify-content-between ']")
for i in minie:
    mini.append(i.text.split("\n")[0])
    maxi.append(i.text.split("\n")[1])
```

```python
glass_com_salary=pd.DataFrame({
    "Company Name":company[:10],
    "No Of Salary":salary[:10],
    "Average Salary":average[:10],
    "Minimum Salary":mini[:10],
    "Maximum Salary":maxi[:10]
})
glass_com_salary
```

Q10.Enter “Laptop” in the search field and then click the search icon.
Then set CPU Type filter to “Intel Core i7” and “Intel Core i9”.
After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. title
2. Ratings
3. Price

```python
amazon=webdriver.Chrome("chromedriver.exe")
amazon.get("https://www.amazon.in/")
```

```python
select1=amazon.find_element_by_id("twotabsearchtextbox")
select1.clear()
select1.send_keys("laptop")
time.sleep(4)
select2=amazon.find_element_by_id("nav-search-submit-button")
select2.click()
```

```python
select3=amazon.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
time.sleep(2)
for i in select3:
    if i.text=="Intel Core i7":
        i.click()
        break
time.sleep(2)
select3=amazon.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
time.sleep(2)
for i in select3:
    if i.text=="Intel Core i9":
        i.click()
        break
```

```python
url=[]
link=amazon.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
for i in link:
    url.append(i.get_attribute("href"))
```

```python
names=[]
ratings=[]
price=[]
for o in url[:10]:
    amazon.get(o)
    try:
        name=amazon.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        names.append(name.text)
    except StaleElementReferenceException:
        names.append("--")
    except NoSuchElementException:
        ratings.append("--")
    try:
        rating=amazon.find_element_by_xpath("//a[@class='a-popover-trigger a-declarative']//i/span")
        ratings.append(rating.get_attribute("innerHTML"))
    except StaleElementReferenceException:
        ratings.append("--")
    except NoSuchElementException:
        ratings.append("--")
    try:
        priz=amazon.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")
        price.append(priz.text) #even by url i cant get exact length
    except StaleElementReferenceException:
        price.append("--")
    except NoSuchElementException:
        price.append("--")
```

```python
laptops=pd.DataFrame({
    "Brands":names,
    "Ratings":ratings,
    "Price":price
})
laptops
```
