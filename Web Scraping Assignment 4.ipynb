```python
import pandas as pd
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import warnings
warnings.filterwarnings("ignore")
```

1.Scrape the details of most viewed videos on YouTube from Wikipedia:

```python
youtube=webdriver.Chrome("chromedriver.exe")
youtube.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos") #open required website
youtube.maximize_window()
youtube.switch_to_window(youtube.window_handles[-1]) #switch tab if unwanted tab open 
time.sleep(4)
row=youtube.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr")      #to get all row
col=youtube.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[1]/td")# to get all columns
value=[]
for r in range(1,len(row)+1): #itterate each value in table by help of rows and columns number
    for c in range(1,len(col)):
        value.append(youtube.find_element_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr["+str(r)+"]/td["+str(c)+"]").text)
youtube.quit()
val=[]
for i in range(5):
    val.append(value[i::5])
columns=["Rank","Name","Artist","Views","Upload date"]
pd.DataFrame(val,index=columns).transpose()    #create pandas DataFrame
```

2.Scrape the details team India’s international fixtures from bcci.tv.

```python
bcci=webdriver.Chrome("chromedriver.exe")
bcci.get("https://www.bcci.tv/.")              #open required website
inte=bcci.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]")
link=inte.find_element_by_xpath(".//*[contains(text(),'Fixtures')]").get_attribute("href") #to reach international fixture page
bcci.get(link)
date=bcci.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']")     # xpath of each data required
des=bcci.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']//p/strong")
place=bcci.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']//p/span")
ser=bcci.find_elements_by_xpath("//div[@class='fixture__format-strip']//span[2]")
dates=[]
mdate=[]
time=[]
match=[]
places=[]
series=[]
for o in date:
    dates.append(o.text.split("\n"))
for i in dates:
    mdate.append(i[:3])
    time.append(i[3])
for i in des:
    match.append(i.text)
for i in place:
    places.append(i.text)
for i in ser:
    series.append(i.text)
bcci.quit()
value=[match,series,places,mdate,time]
columns=["Match title","Series","Place","Date","Time"]
pd.DataFrame(value,index=columns).transpose()          #create pandas DataFrame
```

3.Scrape the details of selenium exception from guru99.com.

4.Scrape the details of State-wise GDP of India from statisticstime.com.

```python
gdp=webdriver.Chrome("chromedriver.exe")
gdp.get("http://statisticstimes.com/")                                                  #open required website
eco=gdp.find_element_by_xpath("//div[@class='dropdown'][2]/div/a[3]").get_attribute("href")  #to reach to economy page
gdp.get(eco)
gdp.find_element_by_xpath("//*[contains(text(),'GDP of Indian states')]").click()
row=gdp.find_elements_by_xpath("//*[@id='table_id']/tbody/tr")                         #to get all row web element
col=gdp.find_elements_by_xpath("//*[@id='table_id']/tbody/tr[1]/td")                   #to get all columns element
values=[]
for r in range(1,len(row)+1):                                  #itterate each value in table by help of rows and columns number
    for c in range(1,len(col)-1):
        values.append(gdp.find_element_by_xpath("//*[@id='table_id']/tbody/tr["+str(r)+"]/td["+str(c)+"]").text)
val=[]
for i in range(6):
    val.append(values[i::6])
gdp.quit()
columns=["Rank","State","GSDP(19-20)","GSDP(18-19)","Share(18-19)","GDP($ billion)"]
pd.DataFrame(val,index=columns).transpose()                                            #create pandas DataFrame
```

5.Scrape the details of trending repositories on Github.com.

```python
hub=webdriver.Chrome("chromedriver.exe")
hub.get("https://github.com/")                                       #open required website
hub.find_element_by_xpath("//*[contains(text(),'Explore')]").click() #to click on the trending option from Explore menu
hub.find_element_by_xpath("//*[contains(text(),'Trending')]").click()
repo=[]
desc=[]
language=[]
cont=[]
top=hub.find_elements_by_xpath("//article")                          #locate each parent tab of data
for r in top:                                                        #exception handling for missing values
    try:
        repo.append(r.find_element_by_xpath(".//h1/a").text)         #locate data xpath form parent tab
    except:
        repo.append("--")
for d in top:
    try:
        desc.append(d.find_element_by_xpath(".//p").text)
    except:
        desc.append("--")
for l in top:
    try:
        language.append(l.find_element_by_xpath(".//span[@itemprop='programmingLanguage']").text)
    except:
        language.append("--")
for c in top:
    try:
        cont.append(c.find_element_by_xpath(".//div[2]/a[1]").text)
    except:
        cont.append("--")
hub.quit()
columns=["Repository title","Repository description","Contributors count","Language used"]
value=[repo,desc,cont,language]
pd.DataFrame(value,columns).transpose()                                #create pandas DataFrame
```

6.Scrape the details of top 100 songs on billiboard.com.

```python
songs=webdriver.Chrome("chromedriver.exe")
songs.get("https://www.billboard.com/")                                         #open required website
link=songs.find_element_by_xpath("//*[contains(text(),'Hot 100')]").get_attribute("href")
songs.get(link)                                           #to click on the charts option then hot 100-page link
names=songs.find_elements_by_xpath("//span[@class='chart-element__information']")   #locate required datas xpath
rate=songs.find_elements_by_xpath("//div[@class='chart-element__metas chart-element__metas--large display--flex flex--y-center']")
sname=[]
aname=[]
lrank=[]
prank=[]
wboard=[]
for i in names:
    sname.append(i.text.split("\n")[0])
    aname.append(i.text.split("\n")[1])
for i in rate:
    lrank.append(i.text.split("\n")[0])
    prank.append(i.text.split("\n")[1])
    wboard.append(i.text.split("\n")[2])
songs.quit()
columns=["Song name","Artist name","Last Week rank","Peak rank","weeks on board"]
val=[sname,aname,lrank,prank,wboard]
pd.DataFrame(val,columns).transpose()                           #create pandas DataFrame
```

7.Scrape the details of Data science recruiters from naukri.com.

```python
import time
naukri=webdriver.Chrome("chromedriver.exe")
naukri.get("https://www.naukri.com/")                                      #open required website
rec=naukri.find_element_by_xpath("//ul[@class='midSec menu']/li[2]/a")
link=rec.get_attribute("href")
naukri.get(link)           #click on the recruiters option and the on the search pane type Data science and click on search
search=naukri.find_element_by_xpath("//div[@class='inpWrap']/input")
search.clear()
search.send_keys("Data science")
naukri.find_element_by_id('qsbFormBtn').click()
naukri.switch_to_window(naukri.window_handles[-1])                         #switch tab if unwanted tab open
time.sleep(4)
names=[]
desig=[]
location=[]
comp=[]
skill=[]
name=naukri.find_elements_by_xpath("//p[@class='highlightable']/a/span")        #locate required data xpath
for n in name:
    names.append(n.text)
des=naukri.find_elements_by_xpath("//p[@class='highlightable']/span[1]")
for d in des:
    desig.append(d.text)
loc=naukri.find_elements_by_xpath("//p[@class='highlightable']")
for l in loc:
    try:                                                                        # to avoid exception if data is missing
        location.append(l.find_element_by_xpath(".//span[2]").text)
    except:
        location.append("--")
com=naukri.find_elements_by_xpath("//p[@class='highlightable']/a[2]")
for c in com:
    comp.append(c.text)
skills=naukri.find_elements_by_xpath("//div[@class='hireSec highlightable']")
for s in skills:
    skill.append(s.text)
columns=["Name","Designation","Company","Skills they hire for","Location"]
value=[names,desig,comp,skill,location]
pd.DataFrame(value,columns).transpose()                                            #create pandas DataFrame
```

9.Scrape the details most watched tv series of all time from imdb.com.

```python
series=webdriver.Chrome("chromedriver.exe")
series.get("https://www.imdb.com/list/ls095964455/")                                  #open required website
names=series.find_elements_by_xpath("//*[@class='lister-item-content']/h3/a")         #locate required web elements by xpath
years=series.find_elements_by_xpath("//*[@class='lister-item-content']/h3/span[2]")
run=series.find_elements_by_xpath("//p[1][@class='text-muted text-small']/span[3]")
gen=series.find_elements_by_xpath("//p[1][@class='text-muted text-small']/span[5]")
vot=series.find_elements_by_xpath("//p[4][@class='text-muted text-small']/span[2]")
alla=series.find_elements_by_xpath("//div[@class='lister-item mode-detail']")         # parent element of rating
name=[]
year=[]
time=[]
genre=[]
rating=[]
vote=[]
for a in names:
    name.append(a.text)
for y in years:
    year.append(y.text)
for r in run:
    time.append(r.text)
for g in gen:
    genre.append(g.text)
for a in alla:
    rating.append(series.find_element_by_xpath(".//div[@class='ipl-rating-widget']").text.split("\n")[0])
for v in vot:
    vote.append(v.text)
series.quit()
columns=["Name","Year span","Genre","Run time","Ratings","Votes"]
value=[name,year,genre,time,rating,vote]
pd.DataFrame(value,columns).transpose()                                                 #create pandas DataFrame
```

10.Details of Datasets from UCI machine learning repositories.

```python
import time
uci=webdriver.Chrome("chromedriver.exe")
uci.get("https://archive.ics.uci.edu/")                                     #open required website
time.sleep(2)
uci.get(uci.find_element_by_xpath("//b[contains(text(),'View ALL Data Sets')]/../..").get_attribute("href")) #to go to the Show All Dataset page
name=uci.find_elements_by_xpath("//td[2]//tbody/tr/td/table")              #locate all rows web element in table
values=[]
for i in range(2,len(name)+2):                                  # itterate by row numbers to locate web element of each value
    val=uci.find_elements_by_xpath("//td[2]//tbody/tr["+str(i)+"]/td")
    for v in val:
        values.append(v.text)
value=[]
for i in range(7):
    value.append(values[i::7])
uci.quit()
columns=["Dataset name","Data type","task","Attribute type","No of instances","No of attribute","Year"]
pd.DataFrame(value,columns).transpose().replace([" "],"-")                   #create pandas DataFrame
```
