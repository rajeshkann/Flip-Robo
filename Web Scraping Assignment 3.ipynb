1. Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.
2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
amazon=webdriver.Chrome("chromedriver.exe")
amazon.get("https://www.amazon.in/")
select1=amazon.find_element_by_id("twotabsearchtextbox")
select1.clear()
input1=input("Enter what you need to scrap : ")
select1.send_keys(input1)
select2=amazon.find_element_by_id("nav-search-submit-button")
select2.click()
url=[]
for o in range(3):
    link=amazon.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a")
    for i in link:
        url.append(i.get_attribute("href"))
    try:
        select3=amazon.find_element_by_xpath("//li[@class='a-last']/a")
        amazon.get(select3.get_attribute("href"))
    except NoSuchElementException:
        break
brand=[]
name=[]
ratings=[]
norating=[]
price=[]
delivery=[]
replace=[]
availabil=[]
other_details=[]
for i in url:
    amazon.get(i)
    try:
        brand.append(amazon.find_element_by_xpath("//*[contains(text(),'Brand')]/../..//*[@class='a-span9']").text)
    except NoSuchElementException:
        try:
            brand.append(amazon.find_element_by_xpath("//*[contains(text(),'Brand')]/..//*[@class='a-size-base prodDetAttrValue']").text)
        except NoSuchElementException:
            brand.append("--")
    except NoSuchElementException:
        brand.append("--")
    try:
        name.append(amazon.find_element_by_id("productTitle").text)
    except NoSuchElementException:
        name.append("--")
    try:
        ratings.append(amazon.find_element_by_xpath("//span[@class='reviewCountTextLinkedHistogram noUnderline']//a/i/span").get_attribute("innerHTML"))
    except NoSuchElementException:
        ratings.append("--")
    try:
        norating.append(amazon.find_element_by_id("acrCustomerReviewText").text)
    except NoSuchElementException:
        norating.append("--")
    try:
        price.append(amazon.find_element_by_xpath("//*[contains(text(),'Price:')]/..//td[2]").text)
    except NoSuchElementException:
        price.append("--")
    try:
        delivery.append(amazon.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b").text)
    except NoSuchElementException:
        delivery.append("--")
    try:
        replace.append(amazon.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']").text)
    except NoSuchElementException:
        replace.append("--")
    try:
        availabil.append(amazon.find_element_by_xpath("//span[contains(text(),'stock')]").get_attribute("innerHTML").replace("\n",""))
    except NoSuchElementException:
        availabil.append("--")
    try:
        other=amazon.find_element_by_id("feature-bullets")
        if other.text.replace("\n","][")=="":
            other_details.append("--")
        else:
            other_details.append(other.text.replace("\n","]["))
    except NoSuchElementException:
        other_details.append("--")
amazon.quit()
columns=("Brand Name","Name of the Product","Rating","No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details","Product URL")
total=(brand,name,ratings,norating,price,replace,delivery,availabil,other_details,url)
pd.DataFrame(total,columns).transpose()
```

3. Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
google=webdriver.Chrome("chromedriver.exe")
google.get("https://images.google.com/?gws_rd=ssl")
google.maximize_window()
list1=["fruits","cars","Machine Learning"]
Fruits=[]
Cars=[]
MachineLearning=[]
for i in list1:
    google.get("https://images.google.com/?gws_rd=ssl")
    select1=google.find_element_by_xpath("//input[@class='gLFyf gsfi']")
    select1.clear()
    select1.send_keys(i)
    select2=google.find_element_by_xpath("//span[@class='z1asCe MZy1Rb']")
    select2.click()
    time.sleep(2)
    try:
        select4=google.find_element_by_xpath("html ")
    except:
        select4=google.find_element_by_xpath("html")
    for e in range(50):
        select4.send_keys(Keys.END)
    time.sleep(3)
    image=google.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
    for o in image:
        if i=="fruits":
            Fruits.append(o.get_attribute("src"))
        if i=="cars":
            Cars.append(o.get_attribute("src"))
        if i=="Machine Learning":
            MachineLearning.append(o.get_attribute("src"))
pd.DataFrame(data=(Fruits[:300],Cars[:300],MachineLearning[:300]),index=list1).transpose()
```

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
flip=webdriver.Chrome("chromedriver.exe")
flip.get("https://www.flipkart.com")
try:
    flip.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()
except NoSuchElementException:
    print("no login")
time.sleep(3)
url=[]
for o in ["Oneplus Nord","pixel 4A"]:
    select1=flip.find_element_by_xpath("//input[@class='_3704LK']")
    select1.send_keys(Keys.CONTROL + "a",Keys.DELETE)
    select1.send_keys(o,Keys.ENTER)
    time.sleep(5)
    link=flip.find_elements_by_xpath("//a[@class='_1fQZEK']")
    for i in link:
        url.append(i.get_attribute("href"))
name=[]
brand=[]
colour=[]
Ram=[]
Rom=[]
primary_camera=[]
secondary_camera=[]
display=[]
display_resolution=[]
processor=[]
processor_cores=[]
battery=[]
price=[]
for i in url:
    flip.get(i)
    flip.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    try:
        names=flip.find_element_by_class_name("B_NuCI").text
        name.append(names)
        brand.append(str(names).split()[0])
    except NoSuchElementException:
        name.append('-')
    try:
        colour.append(flip.find_element_by_xpath("//*[contains(text(),'Color')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        colour.append('-')
    try:
        Ram.append(flip.find_element_by_xpath("//*[contains(text(),'RAM')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        Ram.append('-')
    try:
        Rom.append(flip.find_element_by_xpath("//*[contains(text(),'Internal Storage')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        Rom.append('-')
    try:
        primary_camera.append(flip.find_element_by_xpath("//*[(text()='Primary Camera')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        try:
            primary_camera.append(flip.find_element_by_xpath("//*[contains(text(),'Primary Camera')]/..//*[@class='_21lJbe']").text)
        except NoSuchElementException:
            primary_camera.append('-')
    try:
        secondary_camera.append(flip.find_element_by_xpath("//*[(text()='Secondary Camera')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        try:
            secondary_camera.append(flip.find_element_by_xpath("//*[contains(text(),'Secondary Camera')]/..//*[@class='_21lJbe']").text)
        except NoSuchElementException:
            secondary_camera.append('-')
    try:
        display.append(flip.find_element_by_xpath("//*[contains(text(),'Display Size')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        display.append('-')
    try:
        display_resolution.append(flip.find_element_by_xpath("//*[contains(text(),'Resolution')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        display_resolution.append('-')
    try:
        processor.append(flip.find_element_by_xpath("//*[contains(text(),'Processor Type')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        processor.append('-')
    try:
        processor_cores.append(flip.find_element_by_xpath("//*[contains(text(),'Processor Core')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        processor_cores.append('-')
    try:
        battery.append(flip.find_element_by_xpath("//*[contains(text(),'Battery Capacity')]/..//*[@class='_21lJbe']").text)
    except NoSuchElementException:
        battery.append('-')
    try:
        price.append(flip.find_element_by_xpath("//*[@class='_30jeq3 _16Jk6d']").text)
    except NoSuchElementException:
        price.append('-')
total=(brand,name,colour,Ram,Rom,primary_camera,secondary_camera,display,display_resolution,processor,processor_cores,battery,price,url)
flip_product=pd.DataFrame(total)
phone=flip_product.transpose()
phone.columns=["brand","name","colour","Ram","Rom","primary_camera","secondary_camera","display size","display_resolution","processor","processor_cores","battery","price","url"]
phone
```

7. Write a program to scrap all the available details of top 10 gaming laptops from digit.in.

```python
digit=webdriver.Chrome("chromedriver.exe")
digit.get("https://www.digit.in/")
select1=digit.find_element_by_xpath("//a[contains(text(),'Top 10')]").click()
time.sleep(2)
select2=digit.find_element_by_xpath("//a[contains(text(),'Best Gaming Laptops in India')]")
digit.get(select2.get_attribute("href"))
time.sleep(2)
url=[]
link=digit.find_elements_by_xpath("//*[contains(text(),'Full specs')]")
for i in link:
    url.append(i.get_attribute("data-href"))
names=[]
os=[]
resolution=[]
display_size=[]
ram=[]
processor=[]
rom=[]
battery=[]
batterytype=[]
weight=[]
price=[]
for i in url:
    digit.get(i)
    try:
        names.append(digit.find_element_by_tag_name("h1").text)
    except NoSuchElementException:
        names.append("--")
    try:
        os.append(digit.find_element_by_xpath("//td[contains(text(),'Operating System (With Version)')]/..//td[3]").text)
    except NoSuchElementException:
        os.append("--")
    try:
        resolution.append(digit.find_element_by_xpath("//td[contains(text(),'Resolution')]/..//td[3]").text)
    except NoSuchElementException:
        resolution.append("--")
    try:
        display_size.append(digit.find_element_by_xpath("//td[contains(text(),'Display Size (In Inches)')]/..//td[3]").text)
    except NoSuchElementException:
        display_size.append("--")
    try:
        ram.append(digit.find_element_by_xpath("//td[contains(text(),'Ram Included (In Gb)')]/..//td[3]").text)
    except NoSuchElementException:
        ram.append("--")
    try:
        processor.append(digit.find_element_by_xpath("//td[contains(text(),'Processor Model Name')]/..//td[3]").text)
    except NoSuchElementException:
        processor.append("--")
    try:
        rom.append(digit.find_element_by_xpath("//td[contains(text(),'Storage Drive Capacity')]/..//td[3]").text)
    except NoSuchElementException:
        rom.append("--")
    try:
        battery.append(digit.find_element_by_xpath("//td[contains(text(),'Battery Backup (In Hours)')]/..//td[3]").text)
    except NoSuchElementException:
        battery.append("--")
    try:
        batterytype.append(digit.find_element_by_xpath("//td[contains(text(),'Battery Type')]/..//td[3]").text)
    except NoSuchElementException:
        batterytype.append("--")
    try:
        weight.append(digit.find_element_by_xpath("//td[contains(text(),'Laptop Weight (In Kgs)')]/..//td[3]").text)
    except NoSuchElementException:
        weight.append("--")
    try:
        price.append(digit.find_element_by_xpath("//*[@class='Block-price']/b").text)
    except NoSuchElementException:
        price.append("--")
digit.quit()
columns=("Product Name","OS","Resolution","Display_size","RAM","Processor","ROM","Battery(Back up)","Battery Type","weight","price")
values=(names,os,resolution,display_size,ram,processor,rom,battery,batterytype,weight,price)
pd.DataFrame(values,columns).transpose()
```
